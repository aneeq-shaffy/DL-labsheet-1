3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # Import the 3D toolkit

# Create a figure and a 3D axes object
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the range for X and Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid for X and Y
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X*2 + Y*2

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='inferno') # Use a colormap for better visualization

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D plot of Z = X*2 + Y*2')

plt.show()
```

4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP,
Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the
results using heatmaps. Make sure the values are displaced in the heatmap as well.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset with a different encoding
pokemon_df = pd.read_csv('Pokemon.csv', encoding='latin1')

# Select the features for correlation analysis
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
pokemon_features_df = pokemon_df[features]

# Calculate the Pearson correlation matrix
pearson_corr = pokemon_features_df.corr(method='pearson')

# Calculate the Spearman rank correlation matrix
spearman_corr = pokemon_features_df.corr(method='spearman')

# Visualize the Pearson correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation Matrix')
plt.show()

# Visualize the Spearman rank correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Rank Correlation Matrix')
plt.show()
```

```python
import pandas as pd

# Load the data into a pandas DataFrame using the provided path
df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')

# Display the first few rows of the DataFrame to confirm it loaded correctly
display(df.head())
```

1. Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk,
N_surface and R_avg columns.

```python
# Create a list of the columns you want to keep
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Create a new DataFrame by selecting these columns from the original DataFrame
filtered_df = df[selected_columns]
```

4. Plot the histogram of each of these features in a 1x4 layout.

```python
import matplotlib.pyplot as plt

# Get the list of features to plot
features_to_plot = filtered_df.columns

# Create a figure and a set of subplots with 1 row and 4 columns
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot histogram for each feature
for i, feature in enumerate(features_to_plot):
    axes[i].hist(filtered_df[feature], bins=50, edgecolor='black')
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Show the plot
plt.show()
```
